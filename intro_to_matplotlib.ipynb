```python
x=[1,2,3,4]
y=[4,10,2,1]
plt.plot(x,y)
plt.show(x,y)
```

### Note:

```python
# The pyplot API is generally less-flexible than the object-oreinted API.
# In general, trp to use the object-oriented interface over the pyplot interface.
```

```python
# 1st method 
fig = plt.figure() # creates a figure
ax=fig.add_subplot() # adds some axes
```

```python
# 2nd method 
fig=plt.figure() # creates a figure 
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y) # add some data
plt.show()
```

```python
# 3rd method (recommended )
fig, ax=plt.subplots() 
ax.plot(x,y) #add some data
plt.show()
```

```python
fig, ax=plt.subplots()
ax.plot([],[10,20,30,40])
plt.show()
```

### Making figures with NumPy arrays

```python
x=np.linspace(0,10,100)
ax=plt.subplots(figsize=(10,10))
plt.plot(x,x**2)
plt.scatter(x,np.exp(x))
plt.show()
```

```python
fig, ax=plt.subplots()
ax.scatter(x,np.sin(x))
plt.show()
```

```python
# Make a plot from dictionary
nut_butter_price={'Almond butter':1,
                  'Peanut':20,
                  'Cashew':300}
fig, ax=plt.subplots()
ax.plot(nut_butter_price.keys(),nut_butter_price.values())
ax.scatter(nut_butter_price.keys(),nut_butter_price.values())
plt.show()
```

```python
fig, ax=plt.subplots()
ax.hist(nut_butter_price.keys(),nut_butter_price.values())
plt.show()
```

```python
fig, ax=plt.subplots()
ax.bar(nut_butter_price.keys(),nut_butter_price.values())
plt.show()
```
