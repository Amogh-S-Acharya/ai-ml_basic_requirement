## Dot product example (Nut Butter sales)

```python
np.random.seed(1)
price=np.random.randint(15,size=(1,3))
sales=np.random.randint(20,size=(5,3))
price , sales
```

```python
#total_sales=price.T @ sales 
total_sales=np.dot(sales, price.T, out=None)
total_sales
```

```python
import pandas as pd
pd.DataFrame(price)
```

```python
pd.DataFrame(sales)
```

### Important

```python
pricedf=pd.DataFrame(price,index=['Price'],columns=['Almond Butter','Peanut Butter','Cashew Butter'])
salesdf=pd.DataFrame(sales,index=['Mon','Tue','Wed','Thu','Fri'],columns=['Almond Butter','Peanut Butter','Cashew Butter'])
```

```python
pricedf
```

```python
salesdf
```

```python
salesdf_array_multiplied=salesdf
salesdf_array_multiplied['total_sales']=total_sales
salesdf_array_multiplied
```

```python
total_sales
```

```python
salesdf.shape,pricedf.shape
#daily_sales=salesdf.dot(pricedf.T)
```

```python
daily_sales=salesdf.dot(pricedf.T)
daily_sales
```

```python
salesdf_df_multiplied=salesdf
salesdf_df_multiplied['Total_sales']=daily_sales
salesdf_df_multiplied
```

```python
salesdf_df_multiplied.to_csv('salesdf_df_multiplied.csv')
```
